```python
# B - 1
merged_is_us = pd.merge(cisl_rate_df, cus_rate_df, on=['Month', 'Year'])
merged_is_us.head()
```

```python
# B - 2
merged_is_us['Rate Difference'] = abs(merged_is_us['US rate'] - merged_is_us['ISL rate'])
merged_is_us.head()
```

```python
# B - 3
p = []
i = 0
for isl, us in zip(merged_is_us.iloc[:,0], merged_is_us.iloc[:,5]):
    percentage = (isl - us) / isl * 100
    if percentage >= 1:
        p.append('IH')
    elif percentage <=0:
        p.append('UH')
    else:
        p.append('N')
merged_is_us['Rate Comparison'] = p
merged_is_us.head()
```
